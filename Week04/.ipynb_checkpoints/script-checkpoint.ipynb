# Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm, t
from numpy.random import multivariate_normal
from arch import arch_model
import statsmodels.api as sm
```

```python
# Parameters
S0 = 100 # Initial price
sigma = 0.2 # Volatility
T = 1 # Time horizon
N = 1000 # Number of time steps
dt = T / N # Time step
n = 100000 # Number of simulations

# Classical Brownian Motion
drift_bm = 0
diffusion_bm = sigma
r_bm = np.random.normal(loc=drift_bm*dt, scale=diffusion_bm*np.sqrt(dt), size=(N, n))
P_bm = S0 * np.exp(np.cumsum(r_bm, axis=0))
expected_value_bm = S0 * np.exp(drift_bm*T)
standard_deviation_bm = S0 * np.exp(0.5*diffusion_bm**2*T) * np.sqrt((np.exp(diffusion_bm**2*T)-1))

# Arithmetic Return System
drift_ars = 0.5*sigma**2
diffusion_ars = sigma
r_ars = np.random.normal(loc=drift_ars*dt, scale=diffusion_ars*np.sqrt(dt), size=(N, n))
P_ars = S0 * np.cumprod(1+r_ars, axis=0)
expected_value_ars = S0 * np.exp(drift_ars*T)
standard_deviation_ars = S0 * np.exp((2*drift_ars+diffusion_ars**2)*T) * np.sqrt(np.exp(diffusion_ars**2*T)-1)

# Geometric Brownian Motion
drift_gbm = 0.5*sigma**2
diffusion_gbm = sigma
r_gbm = np.random.normal(loc=(drift_gbm-(diffusion_gbm**2)/2)*dt, scale=diffusion_gbm*np.sqrt(dt), size=(N, n))
P_gbm = S0 * np.exp(np.cumsum(r_gbm, axis=0))
expected_value_gbm = S0 * np.exp(drift_gbm*T)
standard_deviation_gbm = S0 * np.exp(2*drift_gbm*T) * np.sqrt(np.exp(diffusion_gbm**2*T)-1)

# Print results
print("Classical Brownian Motion:")
print(f"Expected value of P_T: {expected_value_bm:.2f}")
print(f"Standard deviation of P_T: {standard_deviation_bm:.2f}")
print("Arithmetic Return System:")
print(f"Expected value of P_T: {expected_value_ars:.2f}")
print(f"Standard deviation of P_T: {standard_deviation_ars:.2f}")
print("Geometric Brownian Motion:")
print(f"Expected value of P_T: {expected_value_gbm:.2f}")
print(f"Standard deviation of P_T: {standard_deviation_gbm:.2f}")
```

```python
# Plot results
plt.figure(figsize=(10, 6))
plt.plot(np.arange(N)*dt, P_bm.mean(axis=1), label="Classical Brownian Motion")
plt.plot(np.arange(N)*dt, P_ars.mean(axis=1), label="Arithmetic Return System")
plt.plot(np.arange(N)*dt, P_gbm.mean(axis=1), label="Geometric Brownian Motion")
plt.axhline(expected_value_bm, color="C0", linestyle="--", label="Expected Value")
plt.axhline(expected_value_ars, color="C1", linestyle="--")
plt.axhline(expected_value_gbm, color="C2", linestyle="--")
plt.legend()
plt.xlabel("Time (years)")
plt.ylabel("Price ($)")
plt.title("Simulated Prices using Different Types of Price Returns")
plt.show()
```

```python
def return_calculate(df, method="DISCRETE"):
    vars = df.columns
    date_column = df['Date'][1]
    n_vars = len(vars)
    vars = [var for var in vars if var != 'Date']
    if n_vars == len(vars):
        raise ValueError(f"date_column: {date_column} not in DataFrame: {vars}")
    n_vars = n_vars - 1
    method="DISCRETE"
    p = df[vars].values
    n = p.shape[0]
    m = p.shape[1]
    p2 = np.empty((n-1, m))

    for i in range(1, n):
        for j in range(m):
            p2[i-1,j] = p[i,j] / p[i-1,j]

    if method.upper() == "DISCRETE":
        p2 = p2 - 1.0
    elif method.upper() == "LOG":
        p2 = np.log(p2)
    else:
        raise ValueError(f"method: {method} must be in (\"LOG\",\"DISCRETE\")")

    dates = df['Date'][1:].values
    out = pd.DataFrame({'Date': dates})
    for i in range(n_vars):
        out[vars[i]] = p2[:,i]
    return out
```

```python
def return_calculate(df, method="DISCRETE"):
    vars = df.columns
    n_vars = len(vars)
    vars = [var for var in vars if var != 'Date']
    n_vars = n_vars - 1
    method="DISCRETE"
    p = df[vars].values
    n = p.shape[0]
    m = p.shape[1]
    p2 = np.empty((n-1, m))

    for i in range(1, n):
        for j in range(m):
            p2[i-1,j] = p[i,j] / p[i-1,j]

    if method.upper() == "DISCRETE":
        p2 = p2 - 1.0
    elif method.upper() == "LOG":
        p2 = np.log(p2)
    else:
        raise ValueError(f"method: {method} must be in (\"LOG\",\"DISCRETE\")")

    out = pd.DataFrame()
    for i in range(n_vars):
        out[vars[i]] = p2[:,i]
    return out
```

```python
# Load data
df = pd.read_csv("DailyPrices.csv")

# Calculate arithmetic returns
returns = return_calculate(df, method="DISCRETE")
returns = returns['META']

# Remove mean from series
returns_centered = returns - returns.mean()
```

```python
# Calculate VaR using normal distribution
var_normal = returns_centered.quantile(0.05)

# Calculate exponentially weighted variance
lambda_094 = 0.94
returns_squared = returns_centered**2
variance_ewm = returns_squared.ewm(alpha=lambda_094, adjust=False).mean()

# Calculate VaR using normal distribution with exponentially weighted variance
var_normal_ewm = var_normal * np.sqrt(variance_ewm.quantile(0.05))
```

```python
# Fit T distribution
params = t.fit(returns_centered.ravel())
t_dist = t(*params)

# Calculate VaR using T distribution
var_t = t_dist.ppf(0.05)


# Fit AR(1) model
model = sm.tsa.ARMA(returns_centered, order=(1,0))
results = model.fit()
ar1_beta = results.params[1]
ar1_sigma = results.resid.std()


# Calculate VaR using AR(1) model
var_ar1 = ar1_beta * returns_centered.iloc[-1] + np.sqrt(ar1_sigma**2 * (1 - ar1_beta**2)) * norm.ppf(0.05)

# Calculate VaR using Historic Simulation
var_hs = -np.percentile(returns_centered, 5)
```

```python
print("VaR (normal):", var_normal)
print("VaR (normal with EW variance):", var_normal_ewm)
print("VaR (T distribution):", var_t)
print("VaR (AR(1)): ", var_ar1)
print("VaR (Historic Simulation):", var_hs)
```

```python
# Load the daily prices and portfolio holdings data
prices_df = pd.read_csv('DailyPrices.csv')
holdings_df = pd.read_csv('Portfolio.csv')
prices_df.set_index('Date', inplace=True)
prices_df = prices_df.pct_change().dropna()
```

```python
# Create an empty dictionary to store the portfolio dataframes
VaR = {}
total_var = 0
# Loop over each portfolio
for portfolio in holdings_df['Portfolio'].unique():
    # Subset the holdings data for the current portfolio
    stocks = holdings_df[holdings_df['Portfolio'] == portfolio]['Stock']
    holdings_subset = holdings_df[holdings_df['Portfolio'] == portfolio]
    holdings_subset.set_index('Stock',inplace = True)
    holdings_subset = holdings_subset.T
    portfolio_ = pd.DataFrame()
    # Calculate the portfolio value for each day and store it
    for i in holdings_subset.columns:
        if i in prices_df.columns:
            portfolio_[i] = prices_df[i] * holdings_df[holdings_df['Stock'] == i]['Holding'].values
    
    total_value = portfolio_.sum(axis=1)
    weights_df = portfolio_.div(total_value, axis=0)
    cov_ewm = portfolio_.ewm(alpha=0.94, adjust=False).cov().dropna()
    # Calculate the portfolio variance as the weighted average of the diagonal elements of the covariance matrix
    portfolio_var = ((cov_ewm.mul(weights_df, axis=0).sum(axis=1) * weights_df.sum(axis=1)).sum())

    # Calculate the portfolio volatility as the square root of the portfolio variance
    portfolio_vol = np.sqrt(portfolio_var)/ np.sqrt(252)

    # Calculate the portfolio VaR
    portfolio_var = (-(weights_df.T.dot(total_value) * portfolio_vol * norm.ppf(0.05))).sum()
    VaR[portfolio] = portfolio_var
    print(f"Portfolio {portfolio} VaR: ${portfolio_var:,.2f}")

for i in VaR.values():
    total_var = total_var + i
print(f"Total VaR: ${total_var:,.2f}")
```

```python
import numpy as np

# Load the daily prices and portfolio holdings data
prices_df = pd.read_csv('DailyPrices.csv')
holdings_df = pd.read_csv('Portfolio.csv')
prices_df.set_index('Date', inplace=True)
prices_df = prices_df.pct_change().dropna()

# Create an empty dictionary to store the portfolio dataframes
VaR = {}
total_var = 0

# Loop over each portfolio
for portfolio in holdings_df['Portfolio'].unique():
    # Subset the holdings data for the current portfolio
    stocks = holdings_df[holdings_df['Portfolio'] == portfolio]['Stock']
    holdings_subset = holdings_df[holdings_df['Portfolio'] == portfolio]
    holdings_subset.set_index('Stock', inplace=True)
    holdings_subset = holdings_subset.T
    portfolio_ = pd.DataFrame()
    
    # Calculate the portfolio value for each day and store it
    for i in holdings_subset.columns:
        if i in prices_df.columns:
            portfolio_[i] = prices_df[i] * holdings_df[holdings_df['Stock'] == i]['Holding'].values
    
    total_value = portfolio_.sum(axis=1)
    weights_df = portfolio_.div(total_value, axis=0)
    
    # Calculate the portfolio VaR using historical simulation
    portfolio_returns = (weights_df * portfolio_).sum(axis=1)
    portfolio_returns = portfolio_returns.dropna()
    portfolio_var = -portfolio_returns.quantile(0.05) * total_value.sum()
    VaR[portfolio] = portfolio_var
    print(f"Portfolio {portfolio} VaR: ${portfolio_var:,.2f}")

for i in VaR.values():
    total_var = total_var + i
print(f"Total VaR: ${total_var:,.2f}")
```
